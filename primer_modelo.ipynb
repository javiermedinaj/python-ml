```python
import pandas as pd
```

```python
datos = pd.read_csv("temperature.csv")
datos.head()
```

```python
import seaborn as sns
```

```python
sns.scatterplot(x="celsius", y="farenheint", data=datos, hue="farenheint", palette="coolwarm")
```

```python
x = datos["celsius"]
y= datos["farenheint"]
```

```python
x
```

```python
y
```

```python
#convertir tipo de dato
x_procesada =  x.values.reshape(-1,1)
y_procesada = y.values.reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
# Crear el modelo
modelo = LinearRegression()
```

```python
#entrenar el modelo
modelo.fit(x_procesada, y_procesada)
```

```python
modelo.predict([[8]])
```

```python
celsius = 6999
prediccion = modelo.predict([[celsius]])
print(f"La temperatura en grados farenheint es: {prediccion}")
```
